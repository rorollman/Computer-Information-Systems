```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({'Month':['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
                    'CityA_High':[33, 37, 44, 56, 67, 75, 79, 78, 71, 60, 50, 38],
                    'CityA_Low':[19, 22, 28, 38, 47, 57, 62, 60, 53, 42, 35, 24],
                    'CityB_High':[54, 59, 67, 76, 83, 91, 93, 94, 84, 75, 63, 55],
                    'CityB_Low':[26, 30, 37, 46, 56, 65, 69, 67, 59, 48, 36, 28]
                    })
data
```

```python
plt.figure(figsize=(20,10))

plt.subplot(3,3,1)
plt.plot(data['Month'],data['CityA_High'], label='City A High')
plt.plot(data['Month'],data['CityA_Low'], 'o', label='City A Low')
plt.title('City A Weather Trend')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.legend(loc='best')
#The graphs follow the same curve

plt.subplot(3,3,2)
plt.plot(data['Month'], data['CityB_High'], '--', label='City B High')
plt.plot(data['Month'], data['CityB_Low'], 'v', label='City B Low')
plt.title('City A Weather Trend')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.legend(loc='best')
#The graphs also folow the same curve

plt.subplot(3,3,3)
plt.bar(data['Month'], data['CityB_High'], label='City B High')
plt.bar(data['Month'], data['CityA_High'], label='City A High')
plt.title('City B vs City A High Temperature')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.legend(loc='best')

plt.subplot(3,3,4)
explode=(0,0,0,0,0,0,0,0,0,.2,0,0)
plt.pie(data['CityA_High'], labels=data['Month'], explode=explode, shadow=True,autopct='%1.1f%%')
plt.title('Pie Chart for City A High Temperature')
#This makes the pie chart title not covered by the x-axis label of the CityA line plot
plt.tight_layout(pad=.5) 
plt.savefig('Q2Charts.pdf')
```
